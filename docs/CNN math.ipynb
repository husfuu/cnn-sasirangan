```python
import numpy as np
from scipy import signal
```

## Convolutional operation

$$conv(I, K) = I * rot180(K)$$

$$ I \ast K = I \star rot180(K)$$

- $\ast$ convolution operation
- $\star$ cros-corelation operation

```python
matrix = np.array([[1, 6, 2],
                   [5, 3, 1],
                   [7, 0, 4]])

kernel = np.array([[1, 2],
                   [-1, 0]])

# rotate kernel 180 degree
kernel = np.rot90(kernel, 2)
```

```python
signal.correlate2d(matrix, kernel, "valid")
```

## Convolutional Layer



- number of input: 1
- depth of input: 3
- number of kernels: 2 
- depth of kernels: 3

The number of matrix outputs is same with the number of kernels
